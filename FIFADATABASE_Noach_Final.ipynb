```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

import scipy.stats as stats
def display_overview(data):
    display(data.info())
    display(data.head())
    display(data.tail())

def visualize_distribution(data, columns):
    sns.set(style="whitegrid")
    fig, axes = plt.subplots(1, len(columns), figsize=(12, 6))
    for i, column in enumerate(columns):
        sns.countplot(x=column, data=data, ax=axes[i])
        axes[i].set_title(f'Value Counts for "{column}"')
    plt.tight_layout()
    plt.show()

def handle_categorical_missing(data, columns):
    for column in columns:
        data[column].fillna('unknown', inplace=True)

def handle_numerical_missing(data, columns):
    for column in columns:
        data[column].fillna(data[column].median(), inplace=True)

def handle_date_column(data, column):
    data[column] = pd.to_datetime(data[column], errors='coerce')
    data[column].fillna(method='ffill', inplace=True)

def visualize_date_counts(data, column):
    plt.figure(figsize=(12, 6))
    data[column].value_counts().sort_index().plot(kind='bar', color='skyblue')
    plt.title(f'Count of occurrences for each date in {column}')
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.show()

# Read the CSV file into a pandas DataFrame
csv_file_path = '/Users/noachmeged/Documents/Ironhack/Project/fifa21_male2.csv'
data = pd.read_csv(csv_file_path)

# Display overview
display_overview(data)

nan_count_per_column = data.isna().sum()


# Select numerical and categorical columns
numerical_columns = data.select_dtypes(include=['number']).columns
categorical_columns = data.select_dtypes(exclude=['number']).columns

# Display numerical and categorical columns
print("Numerical Columns:")
print(numerical_columns)
print("\nCategorical Columns:")
print(categorical_columns)

# Standardize column names
data.columns = data.columns.str.lower().str.replace(" ", "_")

# Handle missing values in categorical columns
handle_categorical_missing(data, ['club', 'position', 'club_logo'])

# Visualize distribution of 'a/w' and 'd/w'
visualize_distribution(data, ['a/w', 'd/w'])

# Handle missing values in 'a/w' and 'd/w' proportionally
for column in ['a/w', 'd/w']:
    distribution = data[column].value_counts(normalize=True)
    data[column] = data[column].apply(lambda x: np.random.choice(distribution.index, p=distribution.values) if pd.isna(x) else x)

# Handle missing values in specific date columns
handle_date_column(data, 'joined')
handle_date_column(data, 'loan_date_end')

# Visualize count of occurrences for each date in 'loan_date_end'
visualize_date_counts(data, 'loan_date_end')

# Handle remaining missing values
for column in data.columns:
    data[column].fillna(data[column].mode().iloc[0], inplace=True)

# Final check for missing values
display_overview(data)

print("Columns with NaN values:")
nan_counts_per_column = data.isna().sum()
for column_name, nan_count in nan_counts_per_column.items():
    print(f"{column_name}: {nan_count}")

for column in data.columns:
    print(f"Value counts for column '{column}':")
    print(data[column].value_counts())
    print("\n")
```

```python
def convert_to_millions(value):
    value = value.replace('€', '')  # Remove currency symbol
    if 'K' in value:
        return float(value.replace('K', '')) / 1000
    elif 'M' in value:
        return float(value.replace('M', ''))
    else:
        return float(value)

# Create a new column 'money_standardized' with values in millions
data['value_in_millions'] = data['value'].apply(convert_to_millions)



#Rank players by market value.


top_10_players = data.sort_values(by='value_in_millions', ascending=False).head(30)

# Display the most valueable players first
print(top_10_players[["name","value_in_millions"]])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Assuming you have a DataFrame named 'data'
# Create separate DataFrames for goalkeepers and other players
goalkeepers = data[data['position'] == 'GK']
other_players = data[data['position'] != 'GK']

# Calculate mean age and confidence interval for goalkeepers
mean_age_goalkeepers = goalkeepers['age'].mean()
ci_goalkeepers = sm.stats.DescrStatsW(goalkeepers['age']).tconfint_mean()

# Calculate mean age and confidence interval for other players
mean_age_other_players = other_players['age'].mean()
ci_other_players = sm.stats.DescrStatsW(other_players['age']).tconfint_mean()

# Display the results
print(f"Mean Age for Goalkeepers: {mean_age_goalkeepers:.2f}")
print(f"Confidence Interval for Goalkeepers: ({ci_goalkeepers[0]:.2f}, {ci_goalkeepers[1]:.2f})")

print(f"\nMean Age for Other Players: {mean_age_other_players:.2f}")
print(f"Confidence Interval for Other Players: ({ci_other_players[0]:.2f}, {ci_other_players[1]:.2f})")

# Plot the results with error bars
sns.barplot(x=['Goalkeepers', 'Other Players'], y=[mean_age_goalkeepers, mean_age_other_players],
            yerr=[(mean_age_goalkeepers - ci_goalkeepers[0], ci_goalkeepers[1] - mean_age_goalkeepers),
                  (mean_age_other_players - ci_other_players[0], ci_other_players[1] - mean_age_other_players)])

plt.title('Mean Age with Confidence Intervals for Goalkeepers and Other Players')
plt.xlabel('Player Type')

```
